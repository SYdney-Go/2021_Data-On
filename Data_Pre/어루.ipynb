```python
data = pd.read_csv("../dataset/어류.csv")
pd.set_option('display.max_columns', None)
data = data["GENUS_KNM"]
data
```

```python
count_list = data.value_counts()
count_list = dict(count_list)
count_list
```

```python
del(count_list['국명없음'])
```

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image

fish = np.array(Image.open('../dataset/fish.png'))

wc = WordCloud(font_path='../dataset/HS유지체.ttf',
               mask = fish,
               background_color = "white",
               contour_width=0.1,
               contour_color='gray',
               colormap = "YlGnBu_r")

wc.generate_from_frequencies(count_list)
wc.to_file('./picture/fish.png')
```
