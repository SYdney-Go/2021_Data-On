```python
data = pd.read_csv("../dataset/관속식물류.csv")
pd.set_option('display.max_columns', None)
data = data["Unnamed: 15"]
data
```

```python
count_list = data.value_counts()
count_list = dict(count_list)
count_list
```

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image

tree = np.array(Image.open('../dataset/tree.png'))

wc = WordCloud(font_path='../dataset/HS유지체.ttf',
               mask = tree,
               background_color = "white",
               contour_width=0.1,
               contour_color='gray',
               colormap = "YlGn")

wc.generate_from_frequencies(count_list)
wc.to_file('./picture/tree.png')
```
